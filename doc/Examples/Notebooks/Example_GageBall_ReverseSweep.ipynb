# Selecting a calibration source for gage ball density

In this example, the density of a gage ball must be calibrated by an external calibration lab. The lab must be selected to meet or exceed density uncertainty requirements. The nominal values are:

Parameter | Value
----------|------
mass $m$     | 86.030 g
diameter $d$  | 22.225 mm
density $\rho$  | 14.967 g/cm$^3$
required $u(\rho)$ | 0.04 g/cm$^3$ (k=2)

Density is determined by the measurement model:

$\rho = \frac{6m}{\pi d^3}$

This is a reverse uncertainty propagation problem, as the uncertainty of the output is known, and we need to determine the uncertainty of the input variables to meet that requirement. However, there are two input variables to work with, so there are two unknowns. To find the combined requirements for measuring mass and diameter, a reverse-sweep calculation can be done. Specify a range of possible $u(d)$ values and solve for the corresponding $u(m)$ that meets the density uncertainty requirement.


```python
from psluncert import reverse, sweeper
```

### Set up the reverse calculation

First, set up a reverse propagation calculation using fixed values for $u(m)$ and $u(d)$. Be aware of units. Here, mass is entered in mg, diameter in microns, and the formula has a conversion factor so the density comes out in g/cm$^3$. Also, the `targetunc` value is entered as a k=1 standard uncertainty.

You could run `.calculate()` on `rcalc` to see the formulas being solved before starting the sweep. 

```python
rcalc = reverse.UncertReverse('rho = 6000000000*m/(pi*d**3)', solvefor='m', targetnom=14.967, targetunc=.02)
rcalc.set_input('m', nom=86030, unc=10)  # mg
rcalc.set_input('d', nom=22225, unc=5);  # um
```

### Sweep the diameter uncertainty

Now, use the `rcalc` set up in the previous cell as the input to a reverse-sweeper. Sweep the diameter uncertainty from 1 to 9. Calling `report_summary()` will show the results of each sweep point in a table and a plot. Any calibration provider that can provide uncertainties below the curve is acceptable to meet the density uncertainty requirement.

```python
s = sweeper.UncertSweepReverse(rcalc)
s.add_sweep_unc('d', list(range(9)))
s.calculate()
s.out.report_summary()
```

Diameter uncertainty is in microns, mass uncertainty in milligrams.
